### Understanding sns library: both axis level and figure level

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
# Scatter plot with regression line
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex')
plt.show()
```

```python
# Bar plot of total bill by day
sns.barplot(x='day', y='total_bill', data=df)
plt.show()
```

```python
# Box plot for distribution of total bill across different days
sns.boxplot(x='day', y='total_bill', data=df)
plt.show()
# Setting a custom color palette
sns.set_palette('pastel')
sns.boxplot(x='day', y='total_bill', data=df)
plt.show()


```

```python
# Violin plot for distribution of total bill by day
sns.violinplot(x='day', y='total_bill', data=df)
plt.show()

# Changing the style of the plot
sns.set_style("whitegrid")
sns.violinplot(x='day', y='total_bill', data=df)
plt.show()

```

```python
import seaborn as sns
import pandas as pd

# Load the tips dataset
df = sns.load_dataset("tips")

# Display the first few rows to understand the dataset
print(df.head())

# Check for non-numeric columns
print(df.dtypes)

# Convert categorical columns into numerical values (if needed)
# For example, 'sex', 'smoker', 'day', 'time' are categorical in the tips dataset.
# Convert them into numerical values using pandas `get_dummies` or `replace`
df['sex'] = df['sex'].replace({'Male': 1, 'Female': 0})
df['smoker'] = df['smoker'].replace({'Yes': 1, 'No': 0})
df['time'] = df['time'].replace({'Dinner': 1, 'Lunch': 0})

# Optionally, use one-hot encoding for the 'day' column (days of the week)
df = pd.get_dummies(df, columns=['day'], drop_first=True)

# Now, calculate the correlation matrix for the numeric columns
correlation_matrix = df.corr()

# Display the correlation matrix
print(correlation_matrix)

corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Pair plot of all numerical columns in the dataset
df = sns.load_dataset("tips")
sns.pairplot(df)
plt.show()
```

Figure-level functions include functions like:

    sns.relplot(): A figure-level function that wraps scatterplot or lineplot.
    sns.catplot(): A figure-level function that wraps categorical plots like barplot and boxplot.
    sns.lmplot(): Used for linear regression plots across multiple facets.
    sns.pairplot(): Used to plot pairwise relationships between variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
df = sns.load_dataset("tips")

# Create a scatter plot with `relplot`
sns.relplot(x="total_bill", y="tip", hue="sex", data=df)

# Show the plot
plt.show()


import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
df = sns.load_dataset('tips')

# Create a figure-level plot
sns.relplot(x='total_bill', y='tip', hue='sex', col='time', data=df)
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a dataset
df = sns.load_dataset("fmri")

# Use relplot to create a line plot
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=df)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
df = sns.load_dataset("tips")

# Create a bar plot with `catplot`
sns.catplot(x="day", y="total_bill", kind="violin", data=df)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
df = sns.load_dataset("tips")

# Create a linear regression plot with `lmplot`
sns.lmplot(x="total_bill", y="tip", data=df)

# Show the plot
plt.show()
```

The full form of lmplot() is "Linear Model Plot." It is specifically designed to show linear relationships between variables with an optional regression line. It fits a linear regression model to the data and adds the regression line to the plot.
